```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("percapita.csv")
```

```python
df.head()
```

### Plotting the Per capita income with correspoding to the year

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (USD)")
plt.title("Per Capita Income of Canada")
plt.plot(df['year'],df['per capita income (US$)']);
```

```python
reg=LinearRegression()
```

```python
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
df.tail()
```

### Predicting the PCI from year 2016 onwards of canada and then plotting it to study and identify the pattern.

```python
test_dict={'year':[2017,2018,2019,2020,2021,2022,2023]}
test_data=pd.DataFrame(test_dict)
predicted_percap=reg.predict(test_data)
```

```python
test_data['per capita income (US$)']=predicted_percap
```

```python
test_data
```

```python
plt.plot(test_data['year'],test_data['per capita income (US$)'])
```

### The model suggests that there will be gradual increase in the per capita of canada. 

#### Combining the year 2017 forth predicted data to the given data upto 2016 to picture the overall progress

```python
final_data=pd.concat([df,test_data],axis=0,ignore_index=True)
```

```python
plt.plot(final_data['year'],final_data['per capita income (US$)'])
plt.plot(test_data['year'],test_data['per capita income (US$)'])
```
